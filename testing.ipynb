```python
no_nan_data = data[~data.bookmaker.isin([u'William Hill', u'Bet365'])]
all_bookmakers = no_nan_data.bookmaker.unique()
```

```python
mean_ratings = no_nan_data.groupby(['bookmaker']).rating.agg(['count', 'mean'])
mean_ratings.sort_values(by=['mean', 'count'], ascending=False)
```

```python
median_ratings = no_nan_data.groupby(['bookmaker']).rating.agg(['count', 'median'])
median_ratings.sort_values(by=['median', 'count'], ascending=False)
```

```python
aggregated_ratings = no_nan_data.groupby(['bookmaker']).rating.agg(['count', 'median', 'mean'])
aggregated_ratings.sort_values(by=['median', 'mean', 'count'], ascending=False)
```
